```python
import numpy as np
from scipy.spatial.distance import squareform
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

d = np.array([[1, 0.4, 0.912, 0.631, 0.7071],
              [0.4, 1, 0, 0.2357, 0.3612],
              [0.912, 0, 1, 0.8137, 0.8452],
              [0.631, 0.2357, 0.8137, 1, 0.5670],
              [0.7071, 0.3612, 0.8452, 0.5670, 1]])

simMat = 1 - d
d = squareform(simMat)
linkage_matrix = linkage(d, method='single')

plt.figure(figsize=(8, 6))
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
import numpy as np
from scipy.spatial.distance import squareform
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

d = np.array([[1, 0.4, 0.912, 0.631, 0.7071],
              [0.4, 1, 0, 0.2357, 0.3612],
              [0.912, 0, 1, 0.8137, 0.8452],
              [0.631, 0.2357, 0.8137, 1, 0.5670],
              [0.7071, 0.3612, 0.8452, 0.5670, 1]])

simMat = 1 - d
d = squareform(simMat)
linkage_matrix = linkage(d, method='complete')

plt.figure(figsize=(8, 6))
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram (Complete Linkage)')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()

```
